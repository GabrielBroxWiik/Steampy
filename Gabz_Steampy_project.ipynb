```python
# Converting minutes into hours and applying to the total playtime column
def convert(minutes):
    hour, min = divmod(minutes, 60)
    return '%02d:%02d' % (hour, min)

def convert_to_hours(minutes):
    return minutes/60

Total_playtime_float = list(map(convert_to_hours, total_playtime))
print(Total_playtime_float)

Total_playtime_in_hours_and_minutes = list(map(convert, total_playtime))
print(Total_playtime_in_hours_and_minutes)


# Driver program
n = 4474				
print(convert(n))
```

```python
# Converting Epoch/Unix time into specific dates
from datetime import datetime
from time import strftime, localtime

def Epochconverter(EpochTime):
    return strftime('%Y-%m-%d %H:%M:%S', localtime(EpochTime))

Last_played_in_Epoch_time = list(map(Epochconverter, last_played))
print(Last_played_in_Epoch_time)
```

```python
# Creating a tuple using the three lists (games, total playtime and last played) and converting to a pandas dataframe
import pandas as pd

Steam_list = list(all_names), list(Total_playtime_float), list(Last_played_in_Epoch_time)
df = pd.DataFrame.from_dict(Steam_list)

# Using the Numpy array and around functions, we can round the total playtime numbers to two decimal places. 
# This is great, because the pandas df.round function did not seem to work for this purpose.
import numpy as np
Total_playtime_float = list(np.around(np.array(Total_playtime_float),2))

# Renaming columns. NB! axis=0 means the rows, but the transpose function (T) switches the rows with columns so axis=0 is for renaming the columns
df = df.set_axis(["Games", "Total playtime in hours and minutes", "Date when last played in Epoch time"])
df.T.sort_values("Total playtime in hours and minutes", axis="rows")
```
